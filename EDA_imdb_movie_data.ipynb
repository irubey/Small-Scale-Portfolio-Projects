```python
#import libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline

#Adjusts the configuration of the plots 
matplotlib.rcParams['figure.figsize'] = (12,8) 

#read in data

df = pd.read_csv(r'C:\Users\isaac\OneDrive\Documents\Portfolio Project\Python Movie Industry Project\movies.csv')
```

```python
#First look at what data is in the set.

df.info()
```

```python
#Is there any missing data?
#Loop through data set and return list of columns with the percentage of values that are null

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
#Drop missing data

df = df.dropna()
```

```python
#Check data set again to see if drop worked

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
#Check data types for our columns

df.dtypes
```

```python
# Alter data types for ease of calculations

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

```

```python
#Add correct year column to dataframe that extracts and copies the year released in order to fix mismatched "year" and "released" values

df['year_correct'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
example_df = pd.DataFrame({'year_correct':df['year_correct'], 'year': df['year'], 'released': df['released']})
example_df.head()

```

```python
#Order movies by gross revenue

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
df
```

```python
# Drop any duplicates

df.drop_duplicates()
```

```python
# Done with cleaning, time to run some analysis!
# Check correlation
# Two predictions:
# 1: Budget may have high correlation with revenue
# 2: Company may have high correlation with revenue

```

```python
#Build scatter plot with movie budget vs gross earnings

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget of Firm')
plt.show()
plt.close()
```

```python
#Plot Budget vs Gross regression plot using seaborn

sns.regplot(x='budget', y='gross', data = df, scatter_kws={'color': 'red'}, line_kws={'color':'blue'})
```

```python
#Look deeper at correlation

df.corr(method='pearson') #Can look at Pearson, Kendall, or Spearman
```

```python
# Looks like there is a high correlation between a movie's budget and it's gross earnings
# The first prediction was correct
```

```python
# Visualize correlation of variables with a heatmap
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
plt.close()
```

```python
# Now I will analyize correlation between production comany and gross earnings
# First create a new dataframe and replace each production company name (and all other non-numerical columns) with a unique number identifyer
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name]= df_numerized[col_name].cat.codes

df_numerized
```

```python
# Now a heatmap can be visualized between the variables
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
plt.close()
```

```python
df_numerized.corr()
```

```python
# To sort through this large correlation matrix, the pairs can be unstacked
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
# Then sorted by numberical value. 
# These are the top 10 values based on pearson correlation
sorted_pairs = corr_pairs.sort_values()

sorted_pairs.head(10)
```

```python
# To find highest correlation between values
high_corr = sorted_pairs[(sorted_pairs) >.5] 

high_corr
```
